```
import numpy as np
from matplotlib import pyplot as plt

y0=3
y=0

g=-9.8
vx=50/11

vy0=0


t1 = (2*(y-y0)/g)**0.5


print (t1)

x=vx*t1

print (x)

#Time frame
t=np.linspace(0,t1,num=200)

#Plotting
x=vx*t
y=y0+vy0*t+(0.5*g)*t**2

#Gr√°fico
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(t, x)  
#ax.plot(y, t)
ax.set_title('Deslocamento Horizontal') 
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.grid()
plt.show()

fig, ax = plt.subplots(figsize=(10, 5))
#ax.plot(t, x)  
ax.plot(t, y)
ax.set_title('Deslocamento Vertical') 
plt.xlabel("t(s)")
plt.ylabel("y(m)")
plt.grid()
plt.show()

```
